# Yocket data for admits 

```python
from selenium import webdriver
import time
from webdriver_manager.chrome import ChromeDriverManager
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

driver = webdriver.Chrome(ChromeDriverManager().install())
driver.get('https://yocket.in/account/login')

username = driver.find_element_by_id("email")
username.send_keys("<Enter user name>")

time.sleep(1)
driver.find_element_by_id("users_login_email").click()
time.sleep(1)

password = driver.find_element_by_id("password")
password.send_keys("<Enter password>")

time.sleep(1)
driver.find_element_by_id("users_login_password").click()

time.sleep(1)
s=input("If Logged in Sucessfully enter k or otherwise login manually then enter k")
```

```python

data=[]
for i in range(1,5):
    driver.get('https://yocket.in/applications-admits-rejects/1855-california-state-university-fresno/2?page='+str(i))
    name_container = driver.find_elements_by_class_name('col-sm-6')
    for i in name_container:
        k =(i.text)
        t=[i for i in k.strip().split("\n") if len(i) is not 0]
        data.append(t)
    time.sleep(1)
```

```python

df= pd.DataFrame(data)
df.rename(columns={0: 'Name', 'newName2': 'University', 1: 'University', 2: 'Year', 3: 'Status',4: 'GRE',5: 'GRE_SCORE',6: 'Eng_test',7:'Test_score',8: 'Undergrad',9: 'Undergrad_score',11: 'work_ex'}, inplace=True)
df=df.dropna()
df
```

## Admit GRE score summary

```python
scores=[]
for i in df['GRE_SCORE']:
    if i in 'N.A.':
        continue
    else:
        scores.append(int(i))
print(sum(scores)/len(scores))
# r.hist(column='GRE_SCORE')
plt.hist(scores, bins = 30)
```

## Admit TOEFL score summary

```python
scores=[]
for i,j in zip(df['Test_score'],df['Eng_test']):
    if i in 'N.A.' and j not in 'TOEFL':
        continue
    else:
        if 'TOEFL' in j:
            scores.append(int(i))
print(sum(scores)/len(scores))
# r.hist(column='GRE_SCORE')
plt.hist(scores, bins=40, range=[75,120])
```
